```python
# Cmd 1 - Crear schema si no existe
spark.sql("CREATE SCHEMA IF NOT EXISTS main.config")
```

```python
# Cmd 2 - Crear tabla main.config.tbl_Entities
from pyspark.sql.types import (
    StructType, StructField, IntegerType, StringType,
    ShortType, TimestampType
)

schema = StructType([
    StructField("IDEntity",      IntegerType(),   False),
    StructField("EntityName",    StringType(),     False),
    StructField("Environment",   StringType(),     True),
    StructField("ADLSaccount",   StringType(),     True),
    StructField("ContainerName", StringType(),     True),
    StructField("SourcePath",    StringType(),     True),
    StructField("FileExtension", StringType(),     True),
    StructField("TargetPath",    StringType(),     True),
    StructField("Notebook",      StringType(),     False),
    StructField("IsActive",      ShortType(),      False),
    StructField("InsertedDate",  TimestampType(),  False),
    StructField("ModifiedDate",  TimestampType(),  False),
])

df = spark.createDataFrame([], schema)

(df.write
   .format("delta")
   .option("mergeSchema", "true")
   .saveAsTable("main.config.tbl_Entities")
)

print("Tabla main.config.tbl_Entities creada exitosamente.")
```
